```python
import numpy as np
from scipy import stats
import statsmodels.api as sm
from scipy.stats import norm, kstest
from scipy.stats import anderson

kstest = stats.kstest(list, 'norm')
print('total data Kolmogorov-Smirnov test: ', kstest)

kstest1 = stats.kstest(list1, 'norm')
print('nondemented data Kolmogorov-Smirnov test: ', kstest1)


kstest2 = stats.kstest(list2, 'norm')
print('demented data Kolmogorov-Smirnov test: ', kstest2)


kstest3 = stats.kstest(list3, 'norm')
print('converted data Kolmogorov-Smirnov test:', kstest3)


list = np.array(list)
loc, scale = norm.fit(list)
n = norm(loc=loc, scale=scale)
kstest = stats.kstest(list, n.cdf)
print('\ntotal data Kolmogorov-Smirnov test comparing with normalized data: ', kstest)

loc, scale = norm.fit(list1)
n = norm(loc=loc, scale=scale)
kstest1 = stats.kstest(list1, n.cdf)
print('nondemented data Kolmogorov-Smirnov test comparing with normalized data: ', kstest1)

loc, scale = norm.fit(list2)
n = norm(loc=loc, scale=scale)
kstest2 = stats.kstest(list2, n.cdf)
print('demented data Kolmogorov-Smirnov test comparing with normalized data: ', kstest2)

loc, scale = norm.fit(list3)
n = norm(loc=loc, scale=scale)
kstest3 = stats.kstest(list3, n.cdf)
print('converted data Kolmogorov-Smirnov test comparing with normalized data:', kstest3)

stat, pvalue = sm.stats.diagnostic.lilliefors(list, 'norm')
stat1, pvalue1 = sm.stats.diagnostic.lilliefors(list1, 'norm')
stat2, pvalue2 = sm.stats.diagnostic.lilliefors(list2, 'norm')
stat3, pvalue3 = sm.stats.diagnostic.lilliefors(list3, 'norm')
print('\ntotal data lilliefors test: ',
      'statistic =', stat, ', pvalue =', pvalue)
print('nondemented data lilliefors test: ',
      'statistic =', stat1, ', pvalue =', pvalue1)
print('demented data lilliefors test: ',
      'statistic =', stat2, ', pvalue =', pvalue2)
print('converted data lilliefors test: ',
      'statistic =', stat3, ', pvalue =', pvalue3)

jb = stats.jarque_bera(list)
print('\ntotal data jarque-bera test: ', jb)
jb1 = stats.jarque_bera(list1)
print('nondemented data jarque-bera test: ', jb1)
jb2 = stats.jarque_bera(list2)
print('demented data jarque-bera test: ', jb2)
jb3 = stats.jarque_bera(list3)
print('converted data jarque-bera test: ', jb3)

ad = anderson(list)
print('\ntotal data anderson-darling test: ', ad)
ad1 = anderson(list1)
print('nondemented anderson-darling test: ', ad1)
ad2 = anderson(list2)
print('demented anderson-darling test: ', ad2)
ad3 = anderson(list3)
print('converted anderson-darling test: ', ad3)
```

```python
from statsmodels.stats.weightstats import ztest as ztest

Z_statics, p_value = ztest(list1, list2, value=0)
print('\n Z-test between nondementeds and dementeds: ',
      'statistic =', Z_statics, ', pvalue =', p_value)
Z_statics, p_value = ztest(list1, list3, value=0)
print('\n Z-test between nondementeds and converteds: ',
      'statistic =', Z_statics, ', pvalue =', p_value)

Z_statics, p_value = ztest(list2, list3, value=0)
print('\n Z-test between dementeds and converteds: ',
      'statistic =', Z_statics, ', pvalue =', p_value)
```

```python
import scipy.stats as stats

print('\nRatio of variances:',np.var(list1)/np.var(list2))
t_statics, p_value = stats.ttest_ind(list1, list2, equal_var=True)
print('t-test between nondementeds and dementeds: ',
      'statistic =', t_statics, ', pvalue =', p_value)

print('\nRatio of variances:',np.var(list1)/np.var(list3))
t_statics, p_value = stats.ttest_ind(list1, list3, equal_var=True)
print('t-test between nondementeds and converteds: ',
      'statistic =', t_statics, ', pvalue =', p_value)

print('\nRatio of variances:',np.var(list2)/np.var(list3))
t_statics, p_value = stats.ttest_ind(list2, list3, equal_var=True)
print('t-test between dementeds and converteds: ',
      'statistic =', t_statics, ', pvalue =', p_value)
```
